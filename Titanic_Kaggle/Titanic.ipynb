```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
train_dataset = pd.read_csv("train.csv")
test_dataset = pd.read_csv("test.csv")
```

```python
train_dataset.head(80)
```

```python
test_dataset.head()
```

```python
columns_target=["Survived"]
columns_train = ["Age", "Pclass", "Sex", "Fare"]
X = train_dataset[columns_train]
Y = train_dataset[columns_target]
```

```python
#Checking for any incompatible data
X['Sex'].isnull().sum()
X['Pclass'].isnull().sum()
X['Fare'].isnull().sum()
X['Age'].isnull().sum()
#Age has 177 columns with no data
```

```python
#We replace the missing values
X['Age'] = X['Age'].fillna(X['Age'].median())
# X['Age'].isnull().sum()
# The missing age data has been replaced by the median of ages
```

```python
#Now we need to replace Male and Female data values
X['Sex'].replace(to_replace='male', value=0, inplace=True)
X['Sex'].replace(to_replace='female', value=1, inplace=True)
X
```

```python
x_test = test_dataset[['Pclass', 'Sex', 'Age','SibSp']]
x_test['Sex'].replace(to_replace='male', value=0, inplace=True)
x_test['Sex'].replace(to_replace='female', value=1, inplace=True)
x_test['Age'] = x_test['Age'].fillna(X['Age'].median())
```

```python
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(y_test, y_pred)
matrix
```

```python
classifier.score(x_test, y_test)
```

```python
pid = test_dataset['PassengerId']
with open('predictions.csv', 'w') as csvfile:
    filewriter = csv.writer(csvfile, delimiter=',', quotechar='|', quoting=csv.QUOTE_MINIMAL)
    filewriter.writerow(['PassengerId', 'Survived'])
    for i, j in zip(pid, y_pred):
        filewriter.writerow([i, j])
```
